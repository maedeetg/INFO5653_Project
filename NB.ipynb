```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
```

#### Song titles

```python
########## SONG TITLES ##########
# Split into training and testing
df_titles = pd.read_csv("lennon_mccartney_clean.csv")
train_titles, test_titles = train_test_split(df_titles, test_size = 0.3)

# Training and testing labels
train_titles_label = train_titles['LABEL']
test_titles_label = test_titles['LABEL']

# Remove labels
train_titles = train_titles.drop("LABEL", axis = 1)
test_titles = test_titles.drop("LABEL", axis = 1)
```

```python
mn_titles = MultinomialNB()
mn_model_titles = mn_titles.fit(train_titles, train_titles_label)
classes = list(mn_model_titles.classes_)
predictions_titles = mn_model_titles.predict(test_titles)

# Print the probabilities
print("The Multinomial NB Model Prediction Probabilities are: \n")
print(mn_model_titles.predict_proba(test_titles).round(3))

# Confusion matrix
mn_titles_confusion_matrix = confusion_matrix(test_titles_label, predictions_titles)
disp = ConfusionMatrixDisplay(confusion_matrix = mn_titles_confusion_matrix, display_labels = classes)
disp.plot(cmap = 'Blues')
plt.title("Song Titles NB")
plt.show()
```
