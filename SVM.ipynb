```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
########## SONG TITLES ##########
# Split into training and testing
df_titles = pd.read_csv("lennon_mccartney_clean.csv")
train_titles, test_titles = train_test_split(df_titles, test_size = 0.3)

# Training and testing labels
train_titles_label = train_titles['LABEL']
test_titles_label = test_titles['LABEL']

# Remove labels
train_titles = train_titles.drop("LABEL", axis = 1)
test_titles = test_titles.drop("LABEL", axis = 1)
```

```python
########## sigmoid C = 1 ##########
SVM_Model_titles_sigmoid = SVC(kernel = 'sigmoid')
SVM_Model_titles_sigmoid.fit(train_titles, train_titles_label)
classes = SVM_Model_titles_sigmoid.classes_

prediction_titles_sigmoid = SVM_Model_titles_sigmoid.predict(test_titles)

# Confusion matrix visualization
SVM_matrix_titles_sigmoid = confusion_matrix(test_titles_label, prediction_titles_sigmoid)
disp = ConfusionMatrixDisplay(confusion_matrix = SVM_matrix_titles_sigmoid, display_labels = classes)
disp.plot(cmap = 'Blues')
plt.title("Song Titles SVM Sigmoid C = 1")
plt.show()
```
