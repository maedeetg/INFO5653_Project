```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
```

```python
########## SONG TITLES ##########
# Split into training and testing
df_titles = pd.read_csv("lennon_mccartney_clean.csv")
train_titles, test_titles = train_test_split(df_titles, test_size = 0.3)

# Training and testing labels
train_titles_label = train_titles['LABEL']
test_titles_label = test_titles['LABEL']

# Remove labels
train_titles = train_titles.drop("LABEL", axis = 1)
test_titles = test_titles.drop("LABEL", axis = 1)
```

```python
dt_titles = DecisionTreeClassifier()
dt_titles = dt_titles.fit(train_titles, train_titles_label)
print(dt_titles.tree_.node_count) # check that we have at least 8 nodes
feature_names_titles = df_titles.columns.values
classes = dt_titles.classes_
```

```python
tree_plot_titles = tree.plot_tree(dt_titles,
                                  feature_names = list(feature_names_titles),  
                                  class_names = list(classes),
                                  filled = True)

plt.savefig("tree_titles.jpg", dpi=300)
```

```python
tree_plot_titles = tree.plot_tree(dt_titles,
                                  feature_names = list(feature_names_titles),  
                                  class_names = list(classes),
                                  filled = True,
                                  max_depth = 3)

plt.savefig("tree_titles_zoom.jpg", dpi=300)
```

```python
# Confusion matrix
predictions_titles = dt_titles.predict(test_titles)
mn_titles_confusion_matrix = confusion_matrix(test_titles_label, predictions_titles)
disp = ConfusionMatrixDisplay(confusion_matrix = mn_titles_confusion_matrix, display_labels = classes)
disp.plot(cmap = 'Blues')
plt.title("Song Titles Decision Tree")
plt.show()
```
